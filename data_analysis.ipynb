```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\rujha\OneDrive\Desktop\teg_project\TEG_data.csv")
pd.set_option('display.max_columns', None)
print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum()) #looking for unexpected NULL and duplicated values
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Material', y='Seebeck_V_per_K')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# Bi2Te3 has highest Seebeck voltage.
# PbTe slightly lower, Skutterudite in between.
```

```python
# delta_T distribution
# Mostly uniform
sns.histplot(df['Delta_T_C'], bins=40, kde=True)
plt.show()
```

```python
# Rload distribution
# uniform
sns.histplot(df['Rload_ohm'], bins=30, kde=True)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
#Rload and Del_t are independent
```

```python
# Histogram for numeric columns
df.hist(figsize=(12,8), bins=40)
plt.xticks(rotation=30)
plt.show()

# Wide range of temprature difference between 50-300 C
# Values of R spread from 2-20 ohm
```

```python
for col in ['Delta_T_C', 'Rload_ohm', 'Seebeck_V_per_K']:
    sns.boxplot(x=df[col])
    plt.xticks(rotation=30)
    plt.title(col)
    plt.show()
```
